```python
import pandas as pd
from xgboost import XGBClassifier 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score ,accuracy_score,classification_report,roc_curve,confusion_matrix,precision_score
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv('./Dataset/Train.csv')
train_data.head()
```

```python
train_data.shape
```

```python
train_data.isnull().sum()
```

```python
train_mode = train_data.X_12.mode()[0]
train_data.fillna(train_mode,inplace=True)
train_data.isnull().sum()
```

```python
data = train_data.drop(['DATE'],axis=1)
data.head()
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
# Function that auto encodes any dataframe column of type category or object.
def dummyEncode(dataset):
        
        columnsToEncode = list(dataset.select_dtypes(include=['category','object']))
        le = LabelEncoder()
        for feature in columnsToEncode:
            try:
                dataset[feature] = le.fit_transform(dataset[feature])
            except:
                print('Error encoding '+feature)
        return dataset
predictors = dummyEncode(data.iloc[:,:-1])
predictors.head()
```

# Stratified Split

```python
from sklearn.model_selection import train_test_split

target = data['MULTIPLE_OFFENSE']

X_train,X_test,Y_train,Y_test = train_test_split(predictors,target, test_size=0.2, random_state=5,stratify=target)
```

# Balancing Data Sets

```python
from imblearn.combine import SMOTETomek
import seaborn as sns
import matplotlib.pyplot as plt


smote = SMOTETomek(random_state=42,sampling_strategy='not majority',n_jobs=-1)
X_sm, Y_sm =  smote.fit_sample(X_train, Y_train)
```

# Tuning the model

```python
# without balancing
xgb_base = XGBClassifier(n_estimators=150, max_depth=4)

xgb_base.fit(X_sm,Y_sm)
Y_pred = xgb_base.predict(X_test)

print('Accuracy is ',xgb_base.score(X_test, Y_test))
print(classification_report(Y_test, Y_pred))
false_positive_rate, true_positive_rate, thresholds = roc_curve(Y_test, Y_pred)

print('ROC_AUC_SCORE is ', roc_auc_score(Y_test, Y_pred))
plt.clf()
plt.plot(false_positive_rate, true_positive_rate)
print(confusion_matrix(Y_test,Y_pred))
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
plt.show()
```

### Depth and child weight

```python
param_test_1 = {
 'max_depth':range(3,15,2),
 'min_child_weight':range(2,10,2)
}

gsearch_1 = GridSearchCV(estimator = xgb_base, 
                         param_grid = param_test_1, scoring='roc_auc',n_jobs=4,iid=False, cv=5)

gsearch_1.fit(X_sm,Y_sm)

print("Best Params:",gsearch_1.best_params_)
print("Best Score:",gsearch_1.best_score_)
```

# Tuning Regularization Parameters

```python
param_test_2 = {
#  'reg_alpha':[1e-5, 1e-2, 0.1, 1, 100]
    'reg_lambda':[1,0, 0.01, 0.05, 0.1, 0.5]
}
gsearch_2 = GridSearchCV(estimator = XGBClassifier(n_estimators=150, max_depth=3,min_child_weight=2), 
                         param_grid = param_test_2, scoring='roc_auc',n_jobs=4,iid=False, cv=5)

gsearch_2.fit(X_sm,Y_sm)

print("Best Params:",gsearch_2.best_params_)
print("Best Score:",gsearch_2.best_score_)
```

### Tuning Learning Rate

```python
param_test_3 = {
                 'learning_rate':[0.001,0.01,0.1,1.0]
                }
gsearch_3 = GridSearchCV(estimator = XGBClassifier(n_estimators=150, max_depth=3,min_child_weight=4,reg_alpha=0),
                         param_grid = param_test_3, scoring='roc_auc',n_jobs=4,iid=False, cv=5)

gsearch_3.fit(X_sm,Y_sm)

print("Best Params:",gsearch_3.best_params_)
print("Best Score:",gsearch_3.best_score_)
```

# Final Model

```python
def xgboost_final(dataframe,target):
    X = dataframe
    y = target

    x_train,x_val,y_train,y_val = train_test_split(X, y, test_size=0.3, random_state=42)

    smote = SMOTETomek(random_state=42,sampling_strategy='not majority',n_jobs=-1)
    X_sm, y_sm =  smote.fit_sample(x_train, y_train)

    model = XGBClassifier(n_estimators=150, max_depth=3,min_child_weight=2)
    model.fit(X_sm, y_sm)
    y_pred = model.predict(x_val)
    
    print('On Validation Data')
    print('Accuracy is ',model.score(x_val, y_val))
    print(classification_report(y_val, y_pred))
    false_positive_rate, true_positive_rate, thresholds = roc_curve(y_val, y_pred)
    #     print(auc(false_positive_rate, true_positive_rate))
    
    print('ROC_AUC_SCORE is ', roc_auc_score(y_val, y_pred))
    plt.clf()
    plt.plot(false_positive_rate, true_positive_rate)
    print(confusion_matrix(y_val,y_pred))
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve')
    plt.show()

    print('On Test Data')
    y_pred = model.predict(X_test)
    print('Accuracy is ',model.score(X_test, Y_test))
    print(classification_report(Y_test, y_pred))
    false_positive_rate, true_positive_rate, thresholds = roc_curve(Y_test, y_pred)
#     print(auc(false_positive_rate, true_positive_rate))

    print('ROC_AUC_SCORE is ', roc_auc_score(Y_test, y_pred))
    plt.clf()
    plt.plot(false_positive_rate, true_positive_rate)
    print(confusion_matrix(Y_test,y_pred))
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.title('ROC curve')
    plt.show()
    
    return model

model = xgboost_final(X_train,Y_train)
```
